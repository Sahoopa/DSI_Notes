### Reusing your code

See if you can use the code you wrote for Jurrasic Park into a function to grab Title, Gross, Distributor, and MPAA rating for each movie on the top 100.  

```python
response = requests.get(top_100_url)
page = response.text
soup = BeautifulSoup(page, "lxml")
```

```python
data = soup.body('div', {'id' : 'body'})[0].find_all('table', {'bgcolor' : '#ffffff'})[0].find_all('font')
headers = data[:9]
data = data[8:]
```

```python
header = []
for row in headers:
    try:
        header.append(row.a.contents)
    except:
        pass
header
```
